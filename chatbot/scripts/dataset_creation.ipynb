```python
import random
import json

# Business details
business_name = "Tractor Spare Parts.lk"
owner_name = "Thilanka Dilshan"
contact_number = "+94 77 123 4567"
business_email = "tractorsparepartslk.c@gmail.com"

# Expanded lists to create richer variations
tractor_brands = [
    "Massey Ferguson", "John Deere", "New Holland", "Sonalika", "Kubota",
    "Mahindra", "Eicher", "TAFE", "Swaraj", "Escorts", "Case IH", "Claas",
    "Deutz-Fahr", "Yanmar", "Ford", "Fendt"
]

models = [
    "240", "5050D", "3630", "DI 60", "4710", "6020", "7200", "265 DI", "275 DI",
    "Arjun Novo 605", "Solis 5015", "TX 1610", "Euro 50", "JIVO 245", "410 DI",
    "Major 60", "5100M", "MF 9500", "NH 5630", "MU5501"
]

parts = [
    "clutch plate", "alternator", "engine oil", "air filter", "diesel filter",
    "fuel injector", "hydraulic pump", "radiator", "starter motor", "head gasket",
    "gearbox", "steering pump", "battery", "tyre", "brake shoe", "oil filter",
    "fan belt", "piston ring", "transmission oil", "exhaust manifold",
    "water pump", "differential", "axle", "seat cushion", "front loader"
]

prices = [
    950, 1500, 4200, 12500, 8500, 11000, 7200, 3600, 5400, 4500,
    9800, 13500, 6500, 8700, 2950, 3990, 7000, 11250, 14200, 3250
]

warranties = [
    "6-month", "1-year", "3-month", "2-year", "18-month", "9-month"
]

locations = [
    "Anuradhapura", "Colombo", "Kandy", "Galle", "Jaffna",
    "Kurunegala", "Matara", "Negombo", "Ratnapura", "Polonnaruwa",
    "Batticaloa", "Trincomalee", "Badulla", "Chilaw", "Vavuniya",
    "Monaragala", "Kalutara", "Puttalam", "Nuwara Eliya", "Hambantota"
]

actions = [
    "buy", "order", "replace", "repair", "inquire about", "check availability of",
    "get price for", "find", "confirm compatibility for", "book"
]

# Generator function
def generate_data_point():
    brand = random.choice(tractor_brands)
    model = random.choice(models)
    part = random.choice(parts)
    price = random.choice(prices)
    warranty = random.choice(warranties)
    location = random.choice(locations)
    action = random.choice(actions)

    # Instruction templates
    instruction_templates = [
        f"Do you have a {part} for {brand} {model}?",
        f"I want to {action} a {part} for my {brand} {model}.",
        f"Is the {part} compatible with {brand} {model}?",
        f"How much is the {part} for {brand} {model}?",
        f"Can you recommend a good {part} for {brand} {model}?",
        f"Do you deliver parts to {location}?",
        f"Where can I get a {part} for {brand} {model}?",
        f"Can I {action} a {part} for {brand} {model} from your store?",
        f"I'm looking to {action} a {part} for {brand} {model}.",
        f"Is the {part} available for {brand} {model} and can you ship to {location}?"
    ]
    instruction = random.choice(instruction_templates)

    # Response templates with phone + email + business branding
    response_templates = [
        f"Yes, we have {part}s compatible with {brand} {model}. The price is Rs. {price} and it comes with a {warranty} warranty. I'm {owner_name} from {business_name}. Feel free to call us at {contact_number} or email us at {business_email}.",
        f"The {part} for {brand} {model} is available at Rs. {price}. You can contact {business_name} at {contact_number} or email {business_email} to place an order.",
        f"Yes, we deliver to {location}. Delivery charges and timelines will be shared once your order is confirmed. For details, call {contact_number} or email {business_email}.",
        f"For {brand} {model}, we recommend our high-quality {part} priced at Rs. {price}. Contact {business_name} on {contact_number} or via email at {business_email} for more info.",
        f"Currently, the {part} for {brand} {model} is out of stock at {business_name}. You can call {contact_number} or email {business_email} to check back next week.",
        f"We have several options for {part}s compatible with {brand} {model}. Prices start from Rs. {price}. I'm {owner_name}. Call {contact_number} or email {business_email} to know more.",
        f"The {part} comes with a {warranty} warranty and costs Rs. {price}. Please contact {business_name} at {contact_number} or email {business_email} if you'd like to proceed.",
        f"We can arrange delivery of the {part} for {brand} {model} to {location}. For arrangements, please call {contact_number} or email us at {business_email}.",
        f"Yes, the {part} fits the {brand} {model}. The price is Rs. {price}. You can reach {business_name} on {contact_number} or email {business_email} to place your order.",
        f"Unfortunately, the {part} for {brand} {model} is not in stock right now at {business_name}. Please call {contact_number} or email {business_email} and we can notify you when it's available."
    ]
    response = random.choice(response_templates)

    return {"instruction": instruction, "response": response}

# Generate dataset
dataset = [generate_data_point() for _ in range(1000000)]

# Save to JSON file
output_path = '/Volumes/KODAK/folder 02/chatbot/data/raw/tractor_parts_chatbot_dataset.json'

with open(output_path, "w", encoding='utf-8') as f:
    json.dump(dataset, f, ensure_ascii=False, indent=2)

print(f"‚úÖ Dataset generated and saved at: {output_path}")
print("Total records:", len(dataset))
```
